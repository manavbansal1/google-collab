Personal Reference doc for Seaborn library

```python
# import seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Seaborn has some builtin datasets

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
# setting a theme for the plot
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips, x = 'total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')
```

```python
# load the iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

ScatterPlot

```python
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
```

```python
sns.scatterplot(data=iris, x='sepal_width', y = 'petal_width', hue='species', style='species')
```

```python
# loading the titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# count plot
sns.countplot(data=titanic, x='class')
```

```python
sns.countplot(data=titanic, x='survived', hue='who')
```

```python
# Bar chart
sns.barplot(data=titanic, x='sex', y='survived', hue='class')
```

```python
sns.distplot(titanic['age'])
```
