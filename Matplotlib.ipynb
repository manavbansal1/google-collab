Refreshing basic Matlotlib

```python
# importing matplotlib
import matplotlib.pyplot as plt
```

```python
# Import other libraries
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print (x)
```

```python
# sin of the numbers
print(y)
```

```python
# cos of the numbers
print(z)
```

```python
# plotting the data
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('sin wave')
plt.show()
```

```python
# plotting the cosine wave
plt.plot(x, z)
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('cos wave')
plt.show()
```

```python
# parabola
x = np.linspace(-10,10,20)
y = (x**2) + (2*x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y = x^2 + 2x')
plt.title('parabola')
plt.show()
```

```python
plt.plot(x, y, 'g.')
plt.xlabel('x')
plt.ylabel('y = x^2 + 2x')
plt.title('parabola')
plt.show()
```

```python
plt.plot(x, y, 'r--')
plt.xlabel('x')
plt.ylabel('y = x^2 + 2x')
plt.title('parabola')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

BAR PLOTS

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', "German"]
poeple = [100,50,40,20,10]
ax.bar(languages, poeple)
plt.xlabel('Languages')
plt.ylabel('Number of people')
plt.title('Popular languages')
plt.show()
```
