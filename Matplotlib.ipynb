Refreshing basic Matlotlib

```python
# importing matplotlib
import matplotlib.pyplot as plt
```

```python
# Import other libraries
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print (x)
```

```python
# sin of the numbers
print(y)
```

```python
# cos of the numbers
print(z)
```

```python
# plotting the data
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('sin wave')
plt.show()
```

```python
# plotting the cosine wave
plt.plot(x, z)
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('cos wave')
plt.show()
```

```python
# parabola
x = np.linspace(-10,10,20)
y = (x**2) + (2*x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y = x^2 + 2x')
plt.title('parabola')
plt.show()
```

```python
plt.plot(x, y, 'g.')
plt.xlabel('x')
plt.ylabel('y = x^2 + 2x')
plt.title('parabola')
plt.show()
```

```python
plt.plot(x, y, 'r--')
plt.xlabel('x')
plt.ylabel('y = x^2 + 2x')
plt.title('parabola')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

BAR PLOTS

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', "German"]
poeple = [100,50,40,20,10]
ax.bar(languages, poeple)
plt.xlabel('Languages')
plt.ylabel('Number of people')
plt.title('Popular languages')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', "German"]
poeple = [100,50,40,20,10]
ax1.pie(poeple, labels = languages, autopct='%1.1f%%')
plt.title('Popular languages')
plt.show()
```

Scatterplot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax2 = fig2.add_axes([0,0,1,1])
ax2.scatter(x,y)
plt.show()
fig3 = plt.figure()
ax3 = fig3.add_axes([0,0,1,1])
ax3.scatter(x,z)
plt.show()

# combined
fig4 = plt.figure()
ax4 = fig4.add_axes([0,0,1,1])
ax4.scatter(x,y, color = 'g')
ax4.scatter(x,z, color = 'r')
plt.show()
```

Building a 3d scatterplot

```python
fig5 = plt.figure()
ax5 = fig5.add_axes([0,0,1,1], projection='3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax5.scatter(x,y,z, c=z, cmap='Blues')
plt.show()
```
